# C-Means
O C-Means tem uma proposta semelhante ao K-Means, porem ele não associa uma instância de maneira absoluta a um cluster, ele faz uma atribuição de maneira **percentual**. Por exemplo, um elemento x pode pertencer 70% a o cluster 1 e 30% ao cluster 2.

```python
#imports
from sklearn import datasets
import numpy as np
from sklearn.metrics import confusion_matrix
import skfuzzy as fuzz 
```

```python
iris = datasets.load_iris()
```

```python
# Algoritmo definindo 3 clusters (c) e passando a matriz transposta (data.T). 
# Os demais params sao obrigatorios, e os valores passados sao os padroes
r = fuzz.cmeans(data=iris.data.T,c=3,m=2,error=0.005,maxiter=1000,init=None)
```

```python
# obter a porcentagem de um registro pertencer ao cluster da posicao 1 na matriz:
previsoes_porcent = r[1]
previsoes_porcent
```

```python
# visualizar as porcentagens de um registro pertencer a cada cluster. Obs.: A soma é 1=100%
for x in range(150): # pq tem-se 150 instancias agrupadas
  print(previsoes_porcent[0][x], previsoes_porcent[1][x], previsoes_porcent[2][x])
```

```python
# fazendo a previsao pegando a maior porcentagem de cada linha
previsoes = previsoes_porcent.argmax(axis=0)
previsoes
```

```python
confusao = confusion_matrix(iris.target,previsoes)
confusao
```
